```python
data = pd.read_csv("../data/보안등2.csv")
```

```python
data['보안등위치명']=data['보안등위치명'].str.replace('-', '')
data['보안등위치명'] = data['보안등위치명'].str.replace(' ', '')
'''
value_to_remove = ['신당동P','','도','대봉지하보도','대신돋','구','삼덕지하보도','대','상중이동 산','대신돋','상중이동산']
data = data[~data['보안등위치명'].isin(value_to_remove)]

data['보안등위치명'] = data['보안등위치명'].str.replace(' ', '')
data['보안등위치명'] = data['보안등위치명'].str.replace(' ', '')
data['보안등위치명']=data['보안등위치명'].str.replace('-', '')

data['보안등위치명'] = data['보안등위치명'].str.replace(r'\s*(읍|면|동|\W)*$', '')

data.iloc[:, 1] = data.iloc[:, 1].str.replace(r'(동|면|읍)', '')
#data['보안등위치명'] = data['보안등위치명'].str.replace(r'\s*(읍|면|동)$', '')
data['보안등위치명'].value_counts()
'''
# Remove trailing spaces, '읍', '면', '동', and other non-alphabetic characters from '보안등위치명'
data['보안등위치명'] = data['보안등위치명'].str.replace(r'\s*(읍|면|동|\W)*$', '')

# Remove spaces and hyphens from '보안등위치명'
data['보안등위치명'] = data['보안등위치명'].str.replace('[\s-]', '')

# Remove specific values from '보안등위치명'
value_to_remove = ['신당동P', '도', '대봉지하보도', '대신돋', '구', '삼덕지하보도', '대', '상중이동산']
data['보안등위치명'] = data['보안등위치명'].replace(value_to_remove, '')

# Remove '동', '면', '읍' from the second column (index 1)
data.iloc[:, 1] = data.iloc[:, 1].str.replace(r'(동|면|읍)', '')

# Display the modified DataFrame
data['보안등위치명'].value_counts()
```

```python
# Group by '보안등위치명' and sum the '설치개수'
result_df = data.groupby('보안등위치명')['설치개수'].sum().reset_index()

# Rename the columns if needed
result_df.columns = ['보안등위치명', '총설치개수']

# Display the result DataFrame
print(result_df)
```
